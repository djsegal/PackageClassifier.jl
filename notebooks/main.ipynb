```julia
using ProgressMeter
using HTTP
using Gumbo

cur_paths = vcat(
    join.(zip(decibans_db.owner, join.(vcat.(decibans_db.package, ".jl"))), "/"),
    join.(zip(general_db.owner, join.(vcat.(general_db.package, ".jl"))), "/")
)

cur_paths = unique(lowercase.(cur_paths))

cur_headers = ["User-Agent" => "GitHub", "Accept" => "application/vnd.github.html"]

@showprogress 1 "Computing..." for cur_path in cur_paths
    cur_package = replace(split(cur_path, "/")[2], ".jl" => "")
    isfile("../data/readme_html/$(cur_package).txt") && continue
        
    readme_url = "https://api.github.com/repos/" * cur_path * "/readme"
    
    readme_url *= "?" * "client_id=" * ENV["CLIENT_ID"]
    readme_url *= "&" * "client_secret=" * ENV["CLIENT_SECRET"]
    
    cur_request = nothing
    
    try 
        cur_request = HTTP.get(
            readme_url, cur_headers,
            userinfo = """$(ENV["CLIENT_ID"])/$(ENV["CLIENT_SECRET"])"""
        )
    catch cur_error
        ( cur_error.status == 404 ) || rethrow(cur_error)
        open("../data/readme_html/$(cur_package).txt", "w") do io ; write(io, "") ; end
        continue
    end
    
    cur_html = String(cur_request.body)
    
    open("../data/readme_html/$(cur_package).txt", "w") do io ; write(io, String(cur_html)) ; end
end
```

```julia
using Languages
using TextAnalysis

@showprogress 1 "Computing..." for cur_path in cur_paths 
    cur_package = replace(split(cur_path, "/")[2], ".jl" => "")
    cur_html = open("../data/readme_html/$(cur_package).txt") do file ; read(file, String) ; end
    
    cur_text = PackageClassifier.custom_text(parsehtml(cur_html))
    cur_text = replace(cur_text, r"\s+" => " ")
    
    open("../data/readme_raw/$(cur_package).txt", "w") do io ; write(io, cur_text) ; end

    stop_words = stopwords(Languages.English())
    
    clean_text = strip(lowercase(cur_text))
    clean_text = replace(clean_text, r"[^a-zA-Z\s]" => " ")
    clean_text = replace(clean_text, r"\s+" => " ")

    split_text = split(clean_text) 
    
    split_text = filter((cur_word) -> length(cur_word) >= 3, split_text)
    split_text = filter((cur_word) -> !(cur_word in stop_words), split_text)
    split_text = map(cur_word -> stem(Stemmer("english"), cur_word), split_text)
    
    clean_text = join(split_text, " ")
    
    open("../data/readme_clean/$(cur_package).txt", "w") do io ; write(io, clean_text) ; end
end
```
